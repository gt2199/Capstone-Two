# ARIMA Model

```python
from statsmodels.tsa.arima_model import ARIMA
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import TimeSeriesSplit
```

## Tune hyperparameters

Due to the large number of data points, we'll use only the year 2013 for tuning hyperparameters (p,q,d). The model is selected based on the Akaike's Information Criteria (AIC).

```python
train_data = data.loc['2013'][['energy']]

ps = [0,1,2,3,4,5]
qs = [0,1,2,3,4,5]
ds = [0,1,2,3,4,5]
minaic = 10000000000
for p in ps:
    for q in qs:
        for d in ds:
            try:
                model = ARIMA(train_data, order=(p,q,d))
                result = model.fit()
                if result.aic < minaic:
                    minaic = result.aic
                    minbic = result.bic
                    optp = p
                    optq = q
                    optd = d
            except:
                continue
print(optp,optq,optd,minaic,minbic)
```

From the optimization above, the optimal p, q, and d are 4, 1, and 5. This means that the model only requres 4 lags in the historical data (4 hours), 1 degree of differencing to account for non-stationary (1 hour), and size 5 moving average window (5 hours).

## Plot predictions

```python
from sklearn.metrics import mean_squared_error
```

Here, we check the model forcasts versus the real data.

```python
train_data = data_test.loc['2017'][['energy']]
plt.rc("figure", figsize=(15,4))
model = ARIMA(train_data, order=(4,1,5))
result = model.fit()
result.plot_predict(start='2017', end='2017')
plt.show()
```

```python
result.summary()
```

```python
result.save('../models/arima.pkl')
```

```python
my_predict = result.predict(start='2017', end='2017',typ='levels')
```

```python
rmse = np.sqrt(mean_squared_error(train_data[1:], my_predict))
print(rmse)
```

# ARIMA Model + features

Here, we can add additional features (exogeneous variables) to our optimized ARIMA model. Since the temperature appears to be correlated with energy based on EDA, only normalized temperature data are added.

## Plot predictions

```python
train_data = data_test.loc['2017'][['energy']]
features_predict = data_test.loc['2017'][['n_temperature']]

plt.rc("figure", figsize=(15,4))
model_x1 = ARIMA(train_data, exog=features_predict, order=(4,1,5))
result_x1 = model_x1.fit()
result_x1.plot_predict(start='2017', end='2017')
plt.title('ARIMA + temperature')
plt.tight_layout()
plt.savefig('forecast_arima.png')
plt.show()
```

```python
result_x1.summary()
```

```python
result_x1.save('../models/arima_x.pkl')
```

```python
my_predict = result_x1.predict(start='2017', end='2017',typ='levels')
```

```python
rmse = np.sqrt(mean_squared_error(train_data[1:], my_predict))
print(rmse)
```

Comparing the ARIMA two models, the original model without additional features has higher AIC than the model with exogeneous feature (106445.557 versus 106071.567). The performance is also reflected on their RMSE were the original model is an RMSE of 188.8, where the additional temperature feature reduces the RMSE to 184.4.

Thus, including the weather data such as temperatures in this case can improve the performance.
